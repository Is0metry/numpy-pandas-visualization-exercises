```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pydataset import data
%matplotlib inline
sns.set(rc={"figure.figsize":(8,6),"figure.dpi":300, 'savefig.dpi':300})
iris = data('iris')
iris.head(10)
```

# Part 1:

Use the iris database to answer the following questions:

## Q1
What does the distribution of petal lengths look like?


```python
sns.axes_style('darkgrid')
sns.histplot(iris['Petal.Length'],kde=True)
```

**The graph appears to have a bimodal distribution with local means around 1.5 and 5**

## Q2
Is there a correlation between petal length and petal width? Use http://guessthecorrelation.com/ as a hint to how we could visually determine if a correlation exists between two numeric columns.

```python
sns.relplot(x=iris['Petal.Length'],y=iris['Petal.Width'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

**There appears to be a strong correlation between petal length and petal width**

## Q3
Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.

```python
sns.relplot(data=iris,x='Sepal.Length',y='Sepal.Width',hue='Species')
```

**It would not be feasible, as there is no significant difference3 between versicolor and virginica**

# Q4
Which features would be best used to predict species?


```python
sns.pairplot(iris,hue='Species')
```

**The best predictors of species would be petal length and petal width**
