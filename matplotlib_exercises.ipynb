# Q1
Use matplotlib to plot the following equation:
$$x^2 - x + 2$$
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-100,101))
y = [x*x - x + 2 for x in x]
plt.plot(x, y)
plt.annotate('Origin',xy=(0,3), xytext=(0,5000),arrowprops=dict(facecolor='blue'))
```

# Q2
## Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
 $$y = \sqrt{x}$$

```python


x1 = [x /10 for x in range(0,100)]
y1 = [math.sqrt(x) for x in x1]
plt.plot(x1,y1)
plt.title('$y = \sqrt{x}$')
plt.xlabel('X',loc='center')
```

$$y = x^3$$

```python
x2 = [x/10 for x in range(-100,101)]
y2 = [x ** 3 for x in x2]
plt.plot(x2,y2)
plt.xlabel('X')
plt.ylabel('Y',rotation=0)
plt.title('$x^3$',loc='center')
```

$$y = 2^x$$

```python
x3 = list(range(-100,101))
x3 = [x/10 for x in x3]
y3 = [2 ** x for x in x3]
plt.plot(x3,y3)
```

$$y = {1 \over{x+1}}$$

```python
x4 = list(range(-200,1))
x4 = [x / 10 for x in x]
y4 = [math.inf if x == -1 else 1/(x+1) for x in x4]

        
plt.plot(x4,y4)
```

# Q3
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(10,10))
plt.suptitle('Mathy Math Math!')
plt.subplot(221)
plt.plot(x1,y1)
plt.title('$\sqrt{x}$')
plt.subplot(222)
plt.plot(x2,y2)
plt.title('$x^3$')
plt.subplot(223)
plt.plot(x3,y3)
plt.title('$2^X$')
plt.subplot(224)
plt.plot(x4,y4)
plt.title('$1/(1 + x)$')
```
