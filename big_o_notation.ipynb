    Curves to graph
    
    *   $ y = 0n + 1 $ and label the curve "O(1)"
    *   $ y = log(n)$ and label the curve "O(log n)"
    *   $y = n$ and label the curve "O(n)"
    *   $y = n * log(n)$ and label it "O(n log n)"
    *   $y = n ^ 2$ and label it "O(n^2)"
    *  $y = 2^n$ and label it "O(2^n)"
    *   $y = n!$ and label it "O(n!)"
    *   $y = n^n$ and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.figure(figsize=(10,10))
plt.title('Big $O$ notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
n = list(range(1,101))
o_1 = [1 for n in n]
plt.plot(n,o_1,label='$O(1)$')
o_2 = [math.log10(n) for n in n]
plt.plot(n,o_2,label='$O(log(n))$')
o_3 = [n for n in n]
plt.plot(n,o_3, label='$O(n)$')
o_4 = [n * math.log10(n) for n in n]
plt.plot(n,o_4,label='$O(nlog(n))$')
o_5 = [n * n for n in n]
plt.plot(n,o_4,label='$O(n^2)$')
o_6 = [2 ** n for n in n]
plt.plot(n,o_6,label='$O(2^{n})$')
o_7 = [math.factorial(n) for n in n]
plt.plot(n,o_7,label='$O(n!)$')
o_8 = [n ** n for n in n]
plt.plot(n,o_8,label='$O(n^n)$')
plt.ylim(-1,100)
plt.legend(loc='best')
```
